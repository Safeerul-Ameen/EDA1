## Data Cleaning
### 1 Handling Missing Values

```python
df.isnull().sum()
```

```python
# Check the proportion of missing values in each column
missing_values = df.isnull().mean() * 100
missing_values.sort_values(ascending=False)
```

**The dataset has missing values in several columns, but the proportion is relatively small (about 0.99% to 0.94%) for those affected**
* LB, DS, DP, MLTV, Width, Tendency, NSP: ~0.99% missing
* AC: ~0.94% missing

```python
# Impute missing values with the median of each column
df_imputed = df.fillna(df.median())

# Verify that there are no more missing values
missing_values_after_imputation = df_imputed.isnull().sum()

missing_values_after_imputation
```

### 2 Handling Outliers

```python
# Set up the matplotlib figure
plt.figure(figsize=(16, 10))

# Generate boxplots for each column
df_imputed.boxplot(grid=False, rot=45, fontsize=10)

# Display the plot
plt.title('Boxplots of Numerical Variables')
plt.show()
```

```python
## visualisatoion of outliers by boxplot
col_nam=list(df_imputed.columns)
fig=plt.figure(figsize=(20,10))
for i in col_nam:
    df.boxplot(column=i,grid=True,
    patch_artist=True,
    boxprops=dict(color='blue', facecolor='lightblue'),
    whiskerprops=dict(color='green', linewidth=2),
    medianprops=dict(color='orange', linewidth=2),
    flierprops=dict(marker='o', color='purple', markersize=5),
    notch=True,
    vert=False)
    plt.show()
```

```python
# percentage of outlier for each column
column=list(df_imputed.columns)
for i in column:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_imputed[i].quantile(0.25)
    Q3 = df_imputed[i].quantile(0.75)

    # Calculate the Interquartile Range (IQR)
    IQR = Q3 - Q1

    # Determine the lower and upper limits
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df_imputed[(df_imputed[i] < lower_limit) | (df_imputed[i] > upper_limit)]

    # Calculate the percentage of outliers
    percentage_outliers = (len(outliers) / len(df_imputed)) * 100

    print(f"Percentage of outliers in {i}: {percentage_outliers:.2f}%")
```

* These percentages indicate that certain columns have a higher concentration of outliers than others, especially:
* FM (16.32%) ,DP (13.36%), ALTV (14.96%), NSP (26.29%)

### Statistical summary 

```python
# Calculate statistical summary for each column in the dataset
summary_stats = df_imputed.describe().T

# Calculate additional measures: IQR and median
summary_stats['IQR'] = summary_stats['75%'] - summary_stats['25%']
summary_stats['median'] = df_imputed.median()

# Select only relevant columns: mean, median, std, IQR
summary_stats = summary_stats[['mean', 'median', 'std', 'IQR']]

# Display the summary
print(summary_stats)
```

Skewness: If the mean is significantly higher or lower than the median, it suggests that the data is skewed.
  * ALTV has positive skewnness. also it has high IQR which means data is more spread out, which could mean more variability in the measurements

### Histogram

```python
import warnings
# Ignore all warnings
warnings.filterwarnings("ignore")
# Set up the figure for histograms
plt.figure(figsize=(16, 20))
for i, column in enumerate(df_imputed.columns, 1):
    plt.subplot(len(df_imputed.columns)//3 + 1, 3, i)
    sns.histplot(df_imputed[column], bins=20, kde=True)
    plt.title(f'Histogram of {column}')
plt.tight_layout()
plt.show();
```
